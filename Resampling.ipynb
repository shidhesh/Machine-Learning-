```python
from sklearn.datasets import make_classification
from sklearn.utils import resample
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_classification(n_samples=1000, n_classes=2, 
                           weights=[0.8, 0.2], random_state=23)
```

```python
data
```

```python
df = pd.DataFrame(data[0], columns=[f"feature_{i}"for i in range(len(data[0][0]))])
```

```python
df.head()
```

```python
df["target"] = data[1]
```

```python
df.head()
```

```python
df["target"].value_counts()
```

```python
majority = df[df["target"]==0]
majority.shape
```

```python
minority = df[df["target"]==1]
minority.shape
```

```python
majority.drop("target", axis=1, inplace=True)
minority.drop("target", axis=1, inplace=True)
```

```python
downsample = resample(majority, 
                      replace=False,
                      n_samples=len(minority), 
                      random_state=53)
```

```python
down_df = pd.concat([minority, downsample])
```

```python
down_df.head()
```

```python
down_df.shape
```

```python
upsample = resample(minority, 
                    replace=True,
                    n_samples=len(majority), 
                    random_state=53)
```

```python
up_df = pd.concat([majority, upsample])
```

```python
up_df.head()
```

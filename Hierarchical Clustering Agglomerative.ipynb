```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
data = np.random.randint(35, 99, 40)
data2 = np.random.randint(35, 99, 40)

X = pd.DataFrame(data, columns=["Col 1"])
X["Col 2"] = data2
```

```python
sch.dendrogram(sch.linkage(X,method="ward"))
plt.axhline(100, linestyle="--", color="red")
plt.show()
```

```python
plt.scatter(X.iloc[:,0], X.iloc[:,1])
plt.show()
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_features=2, n_samples=100, centers=[[2,3], [4,5], [6,7]])
```

```python
data
```

```python
X_tr = data[0]
y = data[1]
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
y_pred = agg.fit_predict(X_tr)
```

```python
new_pred = []
for i in y_pred:
    if i == 1:
        new_pred.append(2)
    elif i == 2:
        new_pred.append(1)
    else:
        new_pred.append(i)
```

```python
new_pred = np.array(new_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y, new_pred)
```

```python
agg1 = AgglomerativeClustering(n_clusters=1)
```

```python
new_pred2 = list(map(lambda x: 1 if x==2 else (2 if x==1 else 0), y_pred))
np.array(new_pred2)
```

```python
accuracy_score(y, new_pred2)
```
